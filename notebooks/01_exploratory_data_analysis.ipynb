```python
import pandas as pd
```

```python
traffic = pd.read_csv("../data/Full_Dataset_Cycling.csv")
```

```python
traffic.shape
```

```python
traffic.head()
```

```python
traffic["Start_Time"] = pd.to_datetime(traffic["Start_Time"])
traffic["End_Time"] = pd.to_datetime(traffic["End_Time"])
```

```python
one_ex = traffic[(traffic["Site_ID"] == "Location_Tag_1")].groupby("End_Time")["Count"].sum().reset_index()
```

```python
one_ex.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style="darkgrid")
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(data=one_ex, x="End_Time", y="Count")
plt.title("Location Tag 1 Historical Traffic Data from 2019 Aug - NOW");
```

```python
# Zoom in on year 2023
one_ex_202324 = one_ex[one_ex["End_Time"].dt.year == 2023]
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(data=one_ex_202324, x="End_Time", y="Count")
plt.title("Location Tag 1 Historical Traffic Data for Year 2023");
```
